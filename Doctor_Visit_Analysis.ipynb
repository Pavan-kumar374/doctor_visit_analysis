**10.Visualize the percentage of people getting government health insurance due to low income,due to old age and also the percentage of people having private health insurance**

```python
label=['yes','no']
Y = df[df['freepoor']=='yes']
N = df[df['freepoor']=='no']
X = [Y.shape[0],N.shape[0]]
plt.figure(figsize=(8,8))
plt.pie(X,labels=label)
plt.title("% of pepole getting governmemt health insurance due to low income")
plt.show()
Y = df[df['private']=='yes']
N = df[df['private']=='no']
X = [Y.shape[0],N.shape[0]]
plt.figure(figsize=(8,8))
plt.pie(X,labels=label)
plt.title("% of pepole having private health insurance")
plt.show()
Y = df[df['freerepat']=='yes']
N = df[df['freerepat']=='no']
X = [Y.shape[0],N.shape[0]]
plt.figure(figsize=(8,8))
plt.pie(X,labels=label)
plt.title("% of pepole getting government health insurance due to old age,disability or veteran status")
plt.show()
```

**11.Plot a horizontal bar chart to analyze the reduced days of activity due to illness based on gender**

```python
db=df.groupby('gender')['reduced'].sum().to_frame().reset_index()
plt.barh(db['gender'],db['reduced'],color = ['red','green'])
plt.xlabel('gender')
plt.ylabel('gender')
plt.show()
```
